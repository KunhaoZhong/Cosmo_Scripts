```python
from astropy.io import fits
import numpy as np
```

```python
hdulist = fits.open('2pt_NG_final_2ptunblind_02_24_21_wnz_redmagic_covupdate.fits')
hdulist.info()
```

```python
covmat = hdulist['COVMAT'].data

with open('cov_unblinded_final.txt', 'w') as outfile:
    for i in range(len(covmat)):
        for j in range(len(covmat[i])):
            outfile.write(str(i) + ' ')
            outfile.write(str(j) + ' ')
            outfile.write(str(covmat[i][j]))
            outfile.write('\n')
```

```python
#NOTE: Not the same length as the nz_lens in cosmolike now
#NOTE: Not in scientific notation, yet

nz_lens_Z_LOW = hdulist['nz_lens'].data.field('Z_LOW')
nz_lens_Z_HIGH = hdulist['nz_lens'].data.field('Z_HIGH')
nz_lens_BIN1 = hdulist['nz_lens'].data.field('BIN1')
nz_lens_BIN2 = hdulist['nz_lens'].data.field('BIN2')
nz_lens_BIN3 = hdulist['nz_lens'].data.field('BIN3')
nz_lens_BIN4 = hdulist['nz_lens'].data.field('BIN4')
nz_lens_BIN5 = hdulist['nz_lens'].data.field('BIN5')


with open('nz_lens_Y3_unblinded_final.txt', 'w') as outfile:
    for i in range(len(nz_lens_Z_LOW)):
        if i!=0:
            outfile.write(str(nz_lens_Z_LOW[i]) + ' ')
            outfile.write(str(nz_lens_BIN1[i])  + ' ')
            outfile.write(str(nz_lens_BIN2[i])  + ' ')
            outfile.write(str(nz_lens_BIN3[i])  + ' ')
            outfile.write(str(nz_lens_BIN4[i])  + ' ')
            outfile.write(str(nz_lens_BIN5[i])  + ' ')
            outfile.write('\n')
        
```

```python
nz_source_Z_LOW = hdulist['nz_source'].data.field('Z_LOW')
nz_source_BIN1  = hdulist['nz_source'].data.field('BIN1')
nz_source_BIN2  = hdulist['nz_source'].data.field('BIN2')
nz_source_BIN3  = hdulist['nz_source'].data.field('BIN3')
nz_source_BIN4  = hdulist['nz_source'].data.field('BIN4')

with open('nz_source_Y3_unblinded_final.txt', 'w') as outfile:
    for i in range(len(nz_source_Z_LOW)):
            outfile.write(str(nz_source_Z_LOW[i]) + ' ')
            outfile.write(str(nz_source_BIN1[i])  + ' ')
            outfile.write(str(nz_source_BIN2[i])  + ' ')
            outfile.write(str(nz_source_BIN3[i])  + ' ')
            outfile.write(str(nz_source_BIN4[i])  + ' ')
            outfile.write('\n')
```

```python
xip = hdulist['xip'].data.field('VALUE')
xim = hdulist['xim'].data.field('VALUE')
gammat = hdulist['gammat'].data.field('VALUE')
wtheta = hdulist['wtheta'].data.field('VALUE')

data = np.concatenate((xip, xim, gammat, wtheta))

with open('Y3_unblinded_final.txt', 'w') as outfile:
    for i in range(len(data)):
        outfile.write(str(i) + ' ')
        outfile.write(str(data[i]))
        outfile.write('\n')
```
